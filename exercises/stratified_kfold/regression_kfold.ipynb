```python
import pandas as pd
import numpy as np

from sklearn import datasets, model_selection
```

```python
X, y = datasets.make_regression(n_samples=15000, n_features=100, n_targets=1)
```

```python
col_names = [f'col_{i}' for i in range(100)]
data_df = pd.DataFrame(X, columns= col_names)
data_df['target'] = y
data_df
```

```python
data_df['kfold'] = -1
```

```python
data_df = data_df.sample(frac=1).reset_index(drop=True)
n_bins = int(np.log2(len(data_df.index)))
n_bins
```

```python
data_df['bins'] = pd.cut(data_df['target'], bins=n_bins, labels=False)
data_df
```

```python
skf = model_selection.StratifiedKFold(n_splits=5)
for f, (train_idx, test_idx) in enumerate(skf.split(X=data_df, y=data_df.bins.values)):
    data_df.loc[test_idx, 'kfold'] = f
data_df
```
